### graph for probability distribution

```python
# ポアソン分布の描画
import math
import matplotlib.pyplot as plt
import numpy as np
def poisson(_lambda, k):
    k = int(k)
    result = (_lambda**k) * (math.exp(-_lambda))  / math.factorial(k)
    return result

_lambda = 6
x =  np.arange(0, 18, 1)
y = [poisson(_lambda,i) for i in x]

plt.bar(x, y, align="center", width=0.4, color="black", label="Poisson λ = {}".format(_lambda))
plt.legend()
plt.savefig("./output/poisson_distribution.png", dpi = 500)
plt.show()
plt.close()
```

```python
# X ~ Poisson(a)
# Y_1, ..., Y_n ~ Poisson(a/n)
# Y = Y_1 + ・・・ + Y_n　とするとき，
# E[X] = E[Y] かつ V[X] = V[Y] となることを確認する．
import numpy as np
import scipy.stats as st
lam = 30
n = 30
size = 10000
x = st.poisson.rvs(lam, size = size)
y_list = []
for i in range(n):
    y_list.append(np.array([st.poisson.rvs(lam/n, size = size)]))
y = np.concatenate([i for i in y_list])
y = np.sum(y, axis=0)
print("E[X]={:.2f}\nV[X]={:.2f}".format(x.mean(), x.var()))
print("E[Y]={:.2f}\nV[Y]={:.2f}".format(y.mean(), y.var()))
```

```python
# 二項分布の描画
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np

n = 10
p = 0.6
x = np.arange(0, n+1)
# P(X=k)
y = st.binom.pmf(x, n, p)

plt.bar(x, y, align="center", width=0.4, color="black", label="Binomial({}, {})".format(n, p))
plt.legend()
plt.show()
```

```python
# ポアソン分布 + 二項分布
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np

n = 10
p = 0.6
lam = 6
x = np.arange(0, 2*n)
# P(X=k)
y_binom = st.binom.pmf(x, n, p)
y_poisson = st.poisson.pmf(x, lam)

plt.bar(x, y_poisson, align="center", width=0.4, color="black", label="Poisson({})".format(lam))
plt.bar(x, y_binom, align="center", width=0.4, color="blue", label="Binomial({}, {})".format(n, p))
plt.legend()
plt.show()
```

```python
# 正規分布の描画
import math
import matplotlib.pyplot as plt
import numpy as np

def gaussian(mu, sigma, x):
    temp = -(x - mu)**2/2/sigma/sigma
    return math.exp(temp) / math.sqrt(2*math.pi) / sigma
    
mu = 40
sigma = 20
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 10000)
y = [gaussian(mu, sigma, xx) for xx in x]
plt.plot(x, y, color = "black", label = "gaussian\nμ = {}\nσ = {}".format(mu, sigma))
plt.legend()
plt.savefig("./output/gaussian_ditribution.png", dpi = 500)
plt.show()
plt.close()
```

```python
# 指数分布
import math
import matplotlib.pyplot as plt
import numpy as np

def exp_dist(lam, x):
    return lam * np.exp(- lam*x) 
    
lam = 10
x = np.linspace(0, 1, 10000)
y = [exp_dist(lam, xx) for xx in x]
plt.plot(x, y, color = "black", label = "exponential({})".format(lam))
plt.legend()
plt.show()
```
